```python
import csv
import math
import re
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# --- parameters ---
SWEEP_DIR = Path('.')           # the notebook lives in the sweep folder
SMOOTH_WINDOW = 50              # rolling-mean window for train curves
USE_LOG_Y = True                # losses span orders of magnitude
AT_STEP = None                  # for lr-vs-loss: None = final, else int step
SAVE_FIGS = False               # set True to overwrite the PNGs in this dir
```

```python
# Discover sweep runs. Filenames look like: run_lr1.03e-01_20260427_135408.csv
LR_RE = re.compile(r'run_lr([0-9.eE+-]+)_\d{8}_\d{6}\.csv$')

def parse_lr(path: Path) -> float | None:
    m = LR_RE.match(path.name)
    return float(m.group(1)) if m else None

runs = []
for p in sorted(SWEEP_DIR.glob('run_lr*.csv')):
    lr = parse_lr(p)
    if lr is not None:
        runs.append((lr, p))

runs.sort(key=lambda x: x[0])
print(f'found {len(runs)} runs:')
for lr, p in runs:
    print(f'  lr={lr:.2e}  {p.name}')
```
